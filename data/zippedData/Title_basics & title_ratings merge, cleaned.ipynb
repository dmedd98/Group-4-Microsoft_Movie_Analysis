```python
import pandas as pd
```

```python
df_title_basics = pd.read_csv('imdb.title.basics.csv.gz')
```

```python
df_title_basics
```

```python
df_title_ratings = pd.read_csv('imdb.title.ratings.csv.gz')
```

```python
df_imdb = df_title_basics.merge(df_title_ratings, left_on = 'tconst', right_on = 'tconst')
```

```python
df_imdb
```

```python
df_imdb.isna().sum()
```

```python
df_imdb.dtypes
```

```python
df_imdb['genres'].fillna('none listed', inplace = True)
df_imdb
```

```python
df_imdb['runtime_minutes'] = df_imdb['runtime_minutes'].fillna(df_imdb['runtime_minutes'].median())
```

```python
df_imdb
```

```python
df_imdb.duplicated('primary_title').value_counts()
```

```python
df_imdb = df_imdb.drop_duplicates(subset = ['primary_title'])
```
